Sales Analysis

```python
import pandas as pd
```

```python
data = {
    "Date": pd.date_range(start="2025-01-01", periods=12, freq="ME"),
    "Product": ["Laptop", "Phone", "Tablet", "Laptop", "Phone", "Laptop",
                "Tablet", "Phone", "Laptop", "Tablet", "Phone", "Laptop"],
    "Quantity": [2, 5, 3, 1, 7, 4, 2, 3, 5, 6, 4, 2],
    "Price": [70000, 25000, 15000, 72000, 26000, 68000,
              14000, 25500, 71000, 14500, 27000, 69000]
}

df = pd.DataFrame(data)
df['Revenue'] = df['Price']*df['Quantity']
print(df)
print()
print("----Monthly sales trend----")
mst = df.groupby("Date")["Revenue"].sum()
print(mst)
print()
print("TOP SELLING PRODUCT")
tsp = df.groupby("Product")["Quantity"].sum()
print(tsp)
top = tsp.max()
top_product = tsp.idxmax()
print(top_product,top)
# Product details dataset
product_details = {
    "Product": ["Laptop", "Phone", "Tablet"],
    "Brand": ["Dell", "Samsung", "Apple"],
    "Category": ["Electronics", "Electronics", "Electronics"]
}

df_products = pd.DataFrame(product_details)
print(df_products)

merge =  df.merge(df_products)
print(merge)
rev = merge.groupby("Category")["Revenue"].sum()
rev_df = rev.reset_index()
rev_df.columns=["category","Total_revenue"]
print(rev_df)

table = pd.pivot_table(df,index="Date",columns="Product",values=["Revenue","Quantity"],aggfunc={"Quantity": "sum", "Revenue": "sum"},fill_value=0)
print(table)
```
