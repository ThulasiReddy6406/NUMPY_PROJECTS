E-commerce Customer Orders  

```python
import pandas as pd
```

```python
customers = {
    "CustomerID": [101, 102, 103, 104, 105],
    "Name": ["Ravi Kumar", "Sita Reddy", "Manoj Singh", "Priya Sharma", "Arjun Das"],
    "City": ["Hyderabad", "Chennai", "Bangalore", "Mumbai", "Delhi"],
    "Segment": ["Retail", "Corporate", "Retail", "Home Office", "Corporate"]
}

orders = {
    "OrderID": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    "CustomerID": [101, 102, 103, 101, 104, 105, 102, 103],
    "OrderDate": pd.date_range(start="2025-08-01", periods=8, freq="D"),
    "Product": ["Laptop", "Phone", "Tablet", "Headphones", "Monitor", "Laptop", "Phone", "Printer"],
    "Quantity": [1, 2, 1, 3, 2, 1, 1, 2],
    "Price": [70000, 25000, 15000, 3000, 12000, 72000, 26000, 15000]
}

df = pd.DataFrame(customers)
print(df)
print()
print()
df_orders = pd.DataFrame(orders)
df_orders["Revenue"] = df_orders["Quantity"] * df_orders["Price"]
print(df_orders)

merge = df.merge(df_orders)
print(merge)
print()
print()
vc = df.value_counts()
print(vc)

print("----")

most_orders = df_orders["CustomerID"].value_counts().reset_index()

print(most_orders)
# Group by customer and calculate total revenue
customer_revenue = df_orders.groupby("CustomerID")["Revenue"].sum().reset_index()

# Sort by revenue in descending order
top_customer = customer_revenue.sort_values(by="Revenue", ascending=False).head(1)

print("Top Revenue Customer:")
print(top_customer)


# revenue_by_customer = df_orders.groupby("CustomerID")["Revenue"].sum().reset_index()
# revenue_by_customer.columns = ["c_id","rev"]
# print(revenue_by_customer)  
# print()
# print()

# top_re = df.loc[df["Name"]["Revenue"].idxmax()]
# print(top_re)
```
