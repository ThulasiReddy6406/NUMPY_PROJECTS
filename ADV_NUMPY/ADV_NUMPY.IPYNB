```python
import numpy as np
import random
```

# NUMPY PROJECTS 


```python
employees = 20
print("TOTAL_EMPLOYES-",employees)
np.random.seed(0)
emp_ids = np.arange(1,employees + 1) 
ages = np.random.randint(21, 61, 20)
dept_ids = np.random.randint(1, 6, 20)
monthly_salary = np.random.randint(25000, 120001, 20)
sales_units = np.random.randint(0, 51, 20)

data = np.column_stack((emp_ids, ages, dept_ids, monthly_salary, sales_units))
print("         === EMPLOYEE DATA ===")
print("------------------------------------------")
print("[      ID    Age   Dept  Salary  Sales]")
print("------------------------------------------")
print(data[0:5])
print("------------------------------------------")
print("AGES OF FIRST 5 EMPLOYESS")
print(data[:5,1])
print("SALARIES OF every 2nd MEMBERS")
print(data[ : : 2,3])
print("SALARY ABOVE 80K")
print(data[monthly_salary > 80000])
print("Employees in Dept 3 with sales > 30")
print(data[(sales_units > 30) & (dept_ids == 3)])
print(f"avg salary",np.average(monthly_salary))
print("max salary",np.max(monthly_salary))
print("total sales",np.sum(sales_units))
print("lowest 3 salaries")
low = np.argsort(data[ : ,3])
print(data[low][:3])
print("top sales")
hig = np.argsort(data[: ,4])[ : : -1]
print(data[hig[ : 3]])
print("YEARLY SALARY")
y_s = data[: ,3]*12
print("ID , YEARLY_SALARY")
print(np.column_stack((data[:,0],y_s)))
performance = np.where(data[:, 4] > 25, "High", "Low")
print("\nPerformance Tags:\n", performance)

unique_depts, dept_counts = np.unique(data[:, 2], return_counts=True)
print("\nUnique Departments:", unique_depts)
print("Employees per Dept:", dept_counts)
print('-----------------------------------------------------------')
print("Total Employees:", len(data))
print("Average Salary:", np.average(monthly_salary))
print("Employees earning > 80k:", np.sum(data[:, 3] > 80000))
print("Top Sales Employee ID:", data[np.argmax(data[:, 4]), 0])
print('-----------------------------------------------------------')
```
